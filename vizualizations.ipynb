```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import scipy.stats as scistats
```

```python
delta = np.linspace(0, 1, num=100)
z = np.linspace(0, 0.25, num=100)
d, z = np.meshgrid(delta, z)

sigma = (scistats.norm.cdf(d)/((1-z)*np.sqrt(2)))**2

plt.figure(figsize=(6,5))
plt.pcolor(d, z, sigma)
plt.colorbar()
plt.show()
```
